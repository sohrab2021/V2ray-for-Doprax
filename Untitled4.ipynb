```python

!pip install datasets langdetect

from datasets import load_dataset
from langdetect import detect
import json

# استفاده از mc4-fa (متن‌های فارسی از دیتاست C4)
ds = load_dataset("allenai/c4", "fa", split="train", streaming=True)

count = 0
max_samples = 10000  # تعداد نمونه‌هایی که می‌خوای ذخیره کنی (قابل تغییره)
output_file = "fa_dataset.jsonl"

with open(output_file, "w", encoding="utf-8") as f:
    for example in ds:
        text = example.get("text", "").strip()
        if len(text) > 100:
            try:
                lang = detect(text)
                if lang == "fa":
                    f.write(json.dumps({"text": text}, ensure_ascii=False) + "\n")
                    count += 1
            except:
                continue
        if count % 500 == 0:
            print(f"{count} نمونه ذخیره شد")
        if count >= max_samples:
            break

print(f"تمام شد! {count} نمونه فارسی در فایل '{output_file}' ذخیره شد.")
```
